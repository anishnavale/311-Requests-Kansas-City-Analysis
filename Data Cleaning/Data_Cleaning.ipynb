```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
# Reading CSV File
df = pd.read_csv("/Users/anishnavale/Desktop/DAMG7250 - BDAG/311_Call_Center_Service_Requests__2007_-_March_2021.csv")
```

```python
df.head()
```

```python
df['SOURCE']=df['SOURCE'].fillna(value='Unknown Source')
df['EXCEEDED EST TIMEFRAME']=df['EXCEEDED EST TIMEFRAME'].fillna(value='Unknown Timeframe')
df['NEIGHBORHOOD']=df['NEIGHBORHOOD'].fillna(value='Unknown Neighborhood')
df['COUNTY']=df['COUNTY'].fillna(value='Unknown Couunty')
df['POLICE DISTRICT']=df['POLICE DISTRICT'].fillna(value='Unknown Police District')
df['CLOSED DATE']=df['CLOSED DATE'].fillna(value='01/01/1900')
df['CLOSED MONTH']=df['CLOSED MONTH'].fillna(value=-99)
df['CLOSED YEAR']=df['CLOSED YEAR'].fillna(value=-99)
df['ZIP CODE']=df['ZIP CODE'].fillna(value=-99)
df['COUNCIL DISTRICT']=df['COUNCIL DISTRICT'].fillna(value='Unknown Council District')
df['STREET ADDRESS']=df['STREET ADDRESS'].fillna(value='Unknown Street Address')
df['30-60-90 Days Open Window']=df['30-60-90 Days Open Window'].fillna(value=1)
df['PARCEL ID NO']=df['PARCEL ID NO'].fillna(value=-99)
df['DAYS TO CLOSE']=df['DAYS TO CLOSE'].fillna(value=-99)
```

```python
df.rename(columns={'CASE ID': 'CASE_ID', 'WORK GROUP': 'WORK_GROUP', 'REQUEST TYPE':'REQUEST_TYPE', 'CREATION DATE':'CREATION_DATE', 'CREATION TIME':'CREATION_TIME', 'CREATION MONTH':'CREATION_MONTH', 'CREATION YEAR':'CREATION_YEAR', 'EXCEEDED EST TIMEFRAME':'EXCEEDED_EST_TIMEFRAME','CLOSED DATE':'CLOSED_DATE', 'CLOSED MONTH':'CLOSED_MONTH', 'CLOSED YEAR':'CLOSED_YEAR', 'DAYS TO CLOSE':'DAYS_TO_CLOSE', 'STREET ADDRESS':'STREET_ADDRESS', 'ADDRESS WITH GEOCODE':'ADDRESS_WITH_GEOCODE', 'ZIP CODE':'ZIP_CODE','COUNCIL DISTRICT':'COUNCIL_DISTRICT', 'POLICE DISTRICT':'POLICE_DISTRICT', 'PARCEL ID NO':'PARCEL_ID_NO','CASE URL':'CASE_URL', '30-60-90 Days Open Window':'30-60-90_Days_Open_Window'}, inplace=True)
```
